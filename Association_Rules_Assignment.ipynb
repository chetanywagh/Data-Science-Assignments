Interview Questions :**bold text**

1) 	What is lift and why is it important in Association rules?

Ans->
1) ift is a metric used to measure the strength of an association rule by comparing the observed support with what would be expected if the items were independent.

2) Agar lift > 1 hai, to iska matlab dono items ek saath expected se zyada baar aa rahe hain – strong positive relation hai.

3) Agar lift = 1 hai, to dono items ka koi khas relation nahi hai – wo bas chance se ek saath aa rahe hain.

4) Agar lift < 1 hai, to items ek saath kam appear hote hain – yaani unka negative relation hai.

5) Lift important isliye hai kyunki ye batata hai ki ek rule sirf frequent nahi, meaningful bhi hai ya nahi.

6) Ye metric specially market basket analysis mein use hoti hai – jaise "jo log bread lete hain, wo butter bhi lete hain ya nahi?"



2).	What is support and Confidence. How do you calculate them?

Ans->
1) Support batata hai ki koi item ya itemset kitni baar poore dataset mein appear hua hai.

2) Confidence batata hai ki agar koi item A aata hai, to uske saath item B aane ki kitni possibility hai.

3) Support se hume pata chalta hai ki koi rule ya combination kitna common hai.

4) Confidence se hume pata chalta hai ki rule kitna reliable hai – yaani A hone par B kitni baar aata hai.

5) In dono ko calculate karne ke liye hum dataset mein items ke appear hone ke frequency ka use karte hain.

3).	What are some limitations or challenges of Association rules mining?

Ans->
1) Too many rules: It can generate thousands of rules, making it hard to find useful ones.

2) Not all rules are meaningful: Some rules may be frequent but not actually useful or interesting.

3) Ignores item importance: It treats all items equally, without knowing which ones are more valuable or relevant.

4) Requires large data: Works best with big datasets; small datasets may not give good or stable rules.

5) Computationally expensive: Finding frequent itemsets can take a lot of time and memory for large datasets.
